```python
import pandas as pd
import numpy as np
import math

df = pd.read_csv("churn_analysis.csv", index_col=0)
df.head()
```

```python
from scipy.stats import chi2_contingency

control_group = df[df["treatment"] == 1]

len(control_group)
```

```python
cg = control_group

#Выбираем штаты
states = pd.unique(cg["state"])
pairs = []

for i in range(len(states) - 1):
    for j in range(i+1, len(states)):
        pairs.append((states[i], states[j]))

len(pairs)
```

```python
cg["churn"] = cg["churn"].map(lambda x: True if x == "True." else False)
cg["churn"]
```

```python
#Q1
q1df = pd.DataFrame({"pairs": pairs})

subtable = []
chi2 = []
i = 0
for pair in pairs:
    a = len(cg[(cg["state"] == pair[0]) & (cg["churn"] == False)])
    b = len(cg[(cg["state"] == pair[0]) & (cg["churn"] == True)])
    c = len(cg[(cg["state"] == pair[1]) & (cg["churn"] == False)])
    d = len(cg[(cg["state"] == pair[0]) & (cg["churn"] == True)])
    subtable.append([[a,b],[c,d]])
    try:
        chi2.append(chi2_contingency([[a,b],[c,d]], correction=False))
    except:
        chi2.append(999)


q1df["subtable"] = subtable
q1df["chi2"] = chi2
    
q1df
```
