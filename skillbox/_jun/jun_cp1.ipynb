### 2.1 Расчет прироста студентов на каждом курсе

```python
### YOUR CODE HERE ###
df = main[["course_title", "progress_id", "start_date", "module_number", "lesson_number", "is_homework"]]

#Находим самые первые домашние работы
df = df[df.is_homework == True]
df["monumber"] = df["module_number"]*100 + df["lesson_number"]

df_min_homeworks = df.groupby("course_title").agg({"monumber": "min"}).reset_index()
min_homeworks = {}

for i, row in df_min_homeworks.iterrows():
    min_homeworks[row["course_title"]] = row["monumber"]
min_homeworks
```

```python
df["month"] = df["start_date"].apply(lambda x: x.month)
df["year"] = df["start_date"].apply(lambda x: x.year)

df_courses = []
for course_title in np.unique(df["course_title"].values):
    df_ = df[(df["course_title"] == course_title) & (df["monumber"] == min_homeworks[course_title])]
    df_ = df_.groupby(["year", "month"]).agg({"progress_id": "nunique"})
    df_dict = df_.to_dict()
    #Проверка на пустые месяцы и годы - через dictionary
    for year in range(2016, 2020):
        for month in range(1, 13):
            if year == 2016 and month < 3: continue
            if (year, month) not in df_dict["progress_id"]:
                df_dict["progress_id"][(year, month)] = 0
            if year == 2019 and month > 6:
                del df_dict["progress_id"][(year, month)]
    df_ = pd.DataFrame(df_dict)
    df_ = df_.reset_index()
    df_["ym"] = df_.apply(lambda x: str(x["level_0"]) + "-" +str(x["level_1"]), axis=1)
    df_ = df_.drop(columns=["level_0", "level_1"])
    df_courses.append((course_title, df_))

len(df_courses)
```

```python
fig, ax = plt.subplots(15, figsize=(30,200))
for i, (title, df) in enumerate(df_courses):
    ax[i].set_title(title, fontsize=20)
    ax[i].tick_params(labelsize=14, labelrotation=90)
    ax[i].set_xlabel("Год-месяц", fontsize=12)
    ax[i].set_ylabel("Прирост студентов за месяц", fontsize=12)
    ax[i].plot(df["ym"], df["progress_id"])
```

```python
ax = plt.figure(figsize=(30,15))
cmap = plt.cm.get_cmap('hsv', len(df_courses))
for i, (title, df) in enumerate(df_courses):
    plt.plot(df["ym"], df["progress_id"], c=cmap(i), label=title)
plt.legend(fontsize=16)
plt.tick_params(labelsize=20, labelrotation=90)
plt.xlabel("Год-месяц", fontsize=20)
plt.ylabel("Прирост студентов за месяц", fontsize=20)
plt.show()
```

### 1.4. Расчет выполняемых домашних заданий по месяцам

```python
df = main[["course_title", "progress_id", "start_date", "finish_date", "status", "is_homework"]]
df = df[df["is_homework"] == True]
df["start_ym"] = df["start_date"].apply(lambda x: x.month + x.year*100)
df["finish_ym"] = df["finish_date"].apply(lambda x: x.month + x.year*100)


df_homeworks = []
for course_title in np.unique(df["course_title"].values):
    df_ = df[df["course_title"] == course_title]
    dct = {"hw_count": {}}
    #Аналогично как в предыдущем задании
    for year in range(2016, 2020):
        for month in range(1, 13):
            cnt = 0
            if year == 2016 and month < 3 or year == 2019 and month > 6: continue
            for p in df_.to_dict(orient='records'):
                #вариант 1 начато в прошлых месяцах и не закончено вообще - СЧИТАЕМ
                if p["start_ym"] <= month + year*100 and p["finish_date"] is None: cnt+=1
                #вариант 2 начато в прошлых месяцах и закончено в этом месяце или в последующих - СЧИТАЕМ
                elif p["start_ym"] <= month + year*100 and\
                   p["finish_ym"] >= month + year*100: cnt+=1
                #иначе не считаем
            dct["hw_count"][(year, month)] = cnt
            
    df_ = pd.DataFrame(dct)
    df_ = df_.reset_index()
    df_["ym"] = df_.apply(lambda x: str(x["level_0"]) + "-" +str(x["level_1"]), axis=1)
    df_ = df_.drop(columns=["level_0", "level_1"])
    df_homeworks.append((course_title, df_))

df_homeworks[0][1]
```

```python
fig, ax = plt.subplots(15, figsize=(30,200))
for i, (title, df) in enumerate(df_homeworks):
    ax[i].set_title(title, fontsize=20)
    ax[i].tick_params(labelsize=14, labelrotation=90)
    ax[i].set_xlabel("Год-месяц", fontsize=12)
    ax[i].set_ylabel("Количество выполненных дз", fontsize=12)
    ax[i].plot(df["ym"], df["hw_count"])
```

```python
ax = plt.figure(figsize=(30,15))
cmap = plt.cm.get_cmap('hsv', len(df_homeworks))
for i, (title, df) in enumerate(df_homeworks):
    plt.plot(df["ym"], df["hw_count"], c=cmap(i), label=title)
plt.legend(fontsize=16)
plt.tick_params(labelsize=20, labelrotation=90)
plt.xlabel("Год-месяц", fontsize=20)
plt.ylabel("Количество выполненных дз", fontsize=20)
plt.show()
```

Равно как и в предыдущем разделе, нагрузка на преподавателей стала расти с 2018 года, с небольшим запаздыванием относительно приема студентов на курсы веб-дизайна мы наблюдаем резкий пик, пришедшийся на конец 2018 года-начало 2019 года, особенно для курса Веб-дизайн с нуля 2.0.
```python
### YOUR CODE HERE ###
df = main[["course_title", "progress_id", "module_number", "start_date", "finish_date", "status", "is_homework"]]
df = df[(df["status"]== "done") & (df["is_homework"] == True)]

df_times = []
for title in df["course_title"].unique():
    df_ = df[(df["course_title"] == title)]
    df_ = df_.groupby(["progress_id", "module_number"]).agg({"finish_date": "max", "start_date": "min"})
    df_["learning_hours"] = df_.apply(lambda x : (x["finish_date"] - x["start_date"]).seconds/3600., axis=1) #считаем в часах
    df_ = df_.groupby("module_number").agg({"learning_hours": ["min","max", "mean","median"]}).reset_index()
    df_["module_number"] = df_["module_number"].astype("int32") #fix
    df_times.append((title, df_))
    
df_times[0][1].head() # для одного курса
```

```python
#По месяцам
df = main[["course_title", "progress_id", "start_date", "finish_date", "status", "is_homework"]]
df = df[(df["status"] == "done") & (df["is_homework"] == True)]
df["month"] = df["finish_date"].apply(lambda x: x.month)
df["year"] = df["finish_date"].apply(lambda x: x.year)

df_monthly = []
for title in df["course_title"].unique():
    df_ = df[df["course_title"] == title]
    df_["hw_hours"] = df_.apply(lambda x : (x["finish_date"] - x["start_date"]).seconds/3600., axis=1)
    df_ = df_.groupby(["month"]).agg({"hw_hours": "median"})    
    df_.reset_index(inplace=True)    
    df_monthly.append((title, df_))
    
df_monthly[0][1].head()
```

```python
ax = plt.figure(figsize=(30,15))
cmap = plt.cm.get_cmap('hsv', len(df_monthly))
for i, (title, df) in enumerate(df_monthly):
    plt.plot(df["month"], df["hw_hours"], c=cmap(i), label=title)
plt.legend(fontsize=16)
#plt.xticks(np.arange(1, 13), 1.0)
plt.tick_params(labelsize=20)
plt.xlabel("Месяц года", fontsize=20)
plt.ylabel("Медианное время выполнения дз, ч", fontsize=20)
plt.xlim([0, 12])
plt.ylim([0, 30])
plt.show()
```

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
### YOUR CODE HERE ###
df = main[["course_title", "module_number", "lesson_number", "status", "is_homework"]]
df = df[df["is_homework"] == True]

df_conv = []
for title in df["course_title"].unique():
    df_ = df[(df["course_title"] == title)]
    conversions = [0.0]
    mnos = list(range(2, np.max(df_["module_number"])+1))
    for mn in mnos:
        df_b = df_[(df_["module_number"] == mn-1)]
        people_done_previous = len(df_b[(df_b["lesson_number"] == np.max(df_b["lesson_number"])) & (df_b["status"] == "done")])
        
        df_a = df_[(df_["module_number"] == mn)]
        people_started_current = len(df_a[df_a["lesson_number"] == np.min(df_a["lesson_number"])])
        
        conv = people_started_current/(people_done_previous + 0.000001) # микро-эпсилон
        if conv > 10.0: conv = 0.0 #ошибочные значения выбрасываем
        conversions.append(conv)
    
    mnos = [1] + mnos
    df_res = pd.DataFrame(np.array([mnos, conversions]).T,\
                          columns=["module_number", "conversion"])
    
    df_conv.append((title, df_res))   
len(df_conv)
```

```python
fig, ax = plt.subplots(15, figsize=(20,150))
for i, (title, df) in enumerate(df_conv):
    ax[i].set_title(title, fontsize=20)
    ax[i].tick_params(labelsize=14)
    ax[i].set_xticks(np.arange(np.min(df["module_number"]), np.max(df["module_number"])+1, 1.0))
    ax[i].set_xlabel("Номер модуля", fontsize=14)
    ax[i].set_ylabel("Конверсия", fontsize=14)
    ax[i].bar(df["module_number"], df["conversion"])
```

Конверсия вычисляется не для всех курсов/модулей. В ряде случаев по условиям формулы получается деление на ноль, эти данные были удалены как выбросы. Можно отметить отсуствие какого-либо явного тренда или формы графиков (ожидалось бы асимптотическое падение?), я полагаю, что это больше связано с построением самого курса, а также интересом студентов или доступностью материала отдельных модулей курса. Можно предположить, что более ровные графики конверсии имеют более популярные курсы, а также **(с учетом собственного опыта)** те курсы, где сдача домашних работ допускается строго последовательно (как оказывается, это скрытый параметр курса на Скиллбоксе).
