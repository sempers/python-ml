```python
from sympy import *
%matplotlib inline

a0, a1, a2, a3, a4, a5 = symbols('a0, a1, a2, a3, a4, a5')

#Пример 1

eqs = [
    -1024*a5 + 256*a4 - 64*a3 + 16*a2 - 4*a1 + a0 +4268,
    -243*a5 + 81*a4 - 27*a3 + 9*a2 -3*a1 + a0 +1227,
    -a5 + a4 - a3 + a2 - a1 + a0 +17,
    a5 + a4 + a3 + a2 + a1 + a0 -17,
    243*a5 + 81*a4 + 27*a3 + 9*a2 +3*a1 + a0 -1227,
    1024*a5 + 256*a4 + 64*a3 + 16*a2 + 4*a1 + a0 - 4268,
]
```

Функция нечетная, имеет три корня ~±0.503, 0. В нуле точка перегиба. Симметричные локальные максимум и минимум в точках, где производная равна нулю, это ~±0.293. Функция возрастает до максимума, пересекая числовую ось, затем убывает, имея перегиб в нуле до локального минимума, затем возрастает до плюс бесконечности.


```python
x = Symbol("x", real=True)
f = 3*x**5 +19*x**3 - 5*x
plot(f, (x, -1, 1))
solveset(f, x)
```

```python
#Пример 2
eqs = [
    -1024*a5 + 256*a4 - 64*a3 + 16*a2 - 4*a1 + a0 +16729,
    -243*a5 + 81*a4 - 27*a3 + 9*a2 -3*a1 + a0 +3999,
    -a5 + a4 - a3 + a2 - a1 + a0 -5,
    a5 + a4 + a3 + a2 + a1 + a0 -1,
    243*a5 + 81*a4 + 27*a3 + 9*a2 +3*a1 + a0 -4005,
    1024*a5 + 256*a4 + 64*a3 + 16*a2 + 4*a1 + a0 - 16735,
]

nonlinsolve(eqs, [a5, a4, a3, a2, a1, a0])
```

Функция не обладает ни свойством четности, ни нечетности, т.к. имеется свободный член, имеет три корня x1 ≈ -1.057, x2 ≈ 0.12, x3 ≈ 0.986. В x2 точка перегиба. Симметричные локальные максимум и минимум в точках (т.к. производная есть уравнение 4 степени, сводящееся к квадратному), где производная равна нулю, это ±0.6656. Функция возрастает до максимума, пересекая числовую ось в x1, затем убывает, имея перегиб и пересечение в x2 до локального минимума, затем возрастает, пересекая ось x в точке x3, до плюс бесконечности.

```python
x = Symbol("x", real=True)
f = 16*x**5 +7*x**3 - 25*x + 3
plot(f, (x, -1.6, 1.6))
solveset(f, x)
```

```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = Symbol('x')
```

```python
#Все коэффициенты подбирал просто руками, без решения неравенств
```

## 1.
### Получите значение MSE меньшее 5.

```python
def print_points_ands_function1(sympy_function):
    def function(x_): return float(sympy_function.subs(x, x_))
    
    points_X = np.array([-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5])
    points_Y = np.array([2, -4, 1, 8, 21, 40, 47, 65, 75, 92])
    plt.xlim(-6, 10)
    plt.ylim(-1, 100)
    
    plt.scatter(points_X, points_Y, c='r')
    x_range = np.linspace(plt.xlim()[0], plt.xlim()[1], num=100)
    function_Y = [function(x_) for x_ in x_range]
    plt.plot(x_range, function_Y, 'b')
    plt.show()
    
    MSE = sum([(points_Y[i] - function(points_X[i]))**2 for i in range(len(points_Y))]) / len(points_Y)
    print(f'MSE = {MSE}')
```

```python
f1 = 48 * x**2 - 128 * x + 83
f1
```

```python
f2 = expand(f1.subs(x, (x+6)/4))
print(f2)
print_points_ands_function1(f2)
```
