```python
from sympy import *
from sympy.plotting import plot, plot3d
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = symbols('x, y')
```

```python
#Пример 1
f = x*(y-5)**2
print("f = ", f)
print("Нули f:", solve(f))
plot3d(f, (x, -5, 5), (y, -5, 5))

ddx = diff(f, x)
ddy = diff(f, y)

print("df/dx: ", ddx)
plot(ddx, (y, -5, 10))
print("Нули df/dx: y = ", solve(ddx))
print("df/dy: ", ddy)
plot3d(ddy,(x, -5, 5), (y, -5, 5))
print("Нули df/dy: ", solve(ddy))
```

```python
#Пример 2
f = (y - exp(1)/20)**2*sin(pi*x)
print("f = ", f)
print("Нули f:", solve(f))
plot3d(f, (x, -5, 5), (y, -5, 5))

ddx = diff(f, x)
ddy = diff(f, y)

print("df/dx: ", ddx)
plot3d(ddx, (x, -2, 2),(y, -0.5, 0.5))
print("Нули df/dx:", solve(ddx))
print("df/dy: ", ddy)
plot3d(ddy,(x, -5, 5), (y, -5, 5))
print("Нули df/dy:", solve(ddx))
```
